```python
set1 = {1,2,3,4,5}
set2 = {4,5,6,7,8}
print(set1 | set2)
print(set1.union(set2))
print(set1 & set2)
print(set1.intersection(set2))
print(set1 - set2)
print(set1.difference(set2))
```

```python
set1 = {1,2,3,4}
set2 = {2,3}
print(set1.issuperset(set2))
print(set2.issubset(set1))
```
