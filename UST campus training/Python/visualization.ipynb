```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
months = ['Jan','Feb','Mar','Apr','May','Jun']
traffic = [150, 2, 100, 220, 250, 310]
plt.plot(months, traffic)
plt.xlabel('Month')
plt.ylabel('Monthly traffic')
plt.title('Monthly website traffic')
plt.savefig("C:/Users/Administrator/Desktop/UST campus training/Plots/image2.png")
plt.show()
```

```python
plt.plot(months, traffic, marker='^', markersize =20, color='green', linestyle='-', linewidth=3)
plt.xlabel('Month')
plt.ylabel('Monthly traffic (in THousands)')
plt.title("Monthly website traffic")
```

```python
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun']
product_a_revenue = [45, 55, 40, 70, 80, 75]
product_b_revenue = [35, 40, 50, 55, 70, 68]

plt.plot(months, product_a_revenue, marker='o', linestyle = '-', color='blue', label = 'Product A')

plt.plot(months, product_b_revenue, marker = 's', linestyle = '--', color='red', label = "Product B")

plt.xlabel("Month")
plt.ylabel("Monthly Revenue (in $1000)")
plt.title("Monthly Revenue Comparison")

plt.legend()
plt.show()
```

```python
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun']

stock1 = [110, 107, 104, 102, 100, 103]
stock2 = [80, 90, 97, 103, 110, 102]

plt.plot(months, stock1, marker = 'v', linestyle = '-.', color='red', label = 'Stock A')
plt.plot(months, stock2, marker = '^', linestyle = '--', color = 'green', label = 'Stock B')
plt.xlabel("Months")
plt.ylabel("Share price")
plt.title("Share price comparison")
plt.legend()
plt.savefig("C:/Users/Administrator/Desktop/UST campus training/Plots/image1.png")
plt.show()
```
