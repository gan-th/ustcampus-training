```python
class Sports:
    def __init__(self, model, color, cost):
        self.__model = model
        self.__color = color
        self.__cost = cost
    def setcost (self, cost):
        self.__cost = cost
    def getcost (self):
        return self.__cost
    def setcolor(self, color):
        self.__color = color

class Shoes(Sports):
    def __init__(self, model, color, cost, brand):
        super().__init__(model, color, cost)
        self.__brand = brand

shoe = Shoes("Shoes", "Black","Rs.4000","ABC")
print(shoe.getcost())
shoe.setcost("Rs.30000")
print(shoe.getcost())
```

```python
# Calculator

def calculator():
    try:
        val1 = int(input("Enter a number"))
        val2 = int(input("Enter the second number: "))
        operator = input("Enter the operator : ( + - * / )")
        if operator == '+':
            print(f"Sum of {val1} and {val2} is {val1+val2}")
        elif operator == '-':
            print(f"Difference of {val1} and {val2} is {val1-val2}")
        elif operator == '*':
            print(f"Product of {val1} and {val2} is {val1*val2}")
        elif operator == '/':
            try:
                print(f"Division of {val1} with {val2} results {val1/val2}")
            except ZeroDivisionError:
                print('Cannot divide a number by zero')
        else:
            print("Operator invalid")
    except ValueError as ve:
        print(" Cannot convert a string to integer")

calculator()
```
