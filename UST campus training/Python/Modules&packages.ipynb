### Python unit testing
## Banking application

```python
global_dict = {}

def deposit(key, value):
    if not isinstance(value,int) :
        return "Can only deposit numerical values"
        
    if value<0:
        return "Can only deposit positive integers"
        
    if key in global_dict:
        global_dict[key] += value
        
    else:
        global_dict[key] = value
    return f"Total balance: {global_dict[key]}"
def withdraw(key, value):
    if not isinstance(value,int) :
        return "Can only withdraw numerical values"
        
    if value<0:
        return "Can only withdraw positive integers"
        
    if key in global_dict:
        if global_dict[key]>=value:
            global_dict[key]-=value
            return f"Withdraw amount {value}"
            
        else:
            return "Not enough balance available"
        return f"Remaining balance: {global_dict[key]}"
    else:
        return "User does not exist"

def check_balance(key):
    if key in global_dict:
        return f"Remaining balance: {global_dict[key]}"
    else:
        return "User does not exist"

```

```python
def test_deposit():
    assert deposit(1,1_000)=="Total balance: 1000"
    assert deposit(1,"abc")=="Can only deposit numerical values"
    assert deposit(1,-1000) =="Can only deposit positive integers"
    print("All deposit test cases passed ")

def test_withdraw():
    deposit(2,1000)
    assert withdraw(2,500)=="Withdraw amount 500"
    assert withdraw(2,'abc')=="Can only withdraw numerical values"
    assert withdraw(2,1000)=="Not enough balance available"
    assert withdraw(2,-1000) =="Can only withdraw positive integers"
    print("All withdraw test cases passed")

def test_check_balance():
    deposit(3,1000)
    assert check_balance(3) == "Remaining balance: 1000"
    assert check_balance(100) == "User does not exist"
    print("Check balance test cases passed")
    
test_deposit()
test_withdraw()
test_check_balance()
```
