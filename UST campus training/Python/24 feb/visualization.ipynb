### Stack Plot

```python
quaters = ['Q1', 'Q2', 'Q3', 'Q4']
electronics = [10000, 12000, 11000, 10500]
clothing = [5000, 6000, 7500, 8000]
home_appliances = [3000, 7500, 9000, 7000]

plt.figure(figsize=(10, 6))
plt.stackplot(quaters, electronics, clothing, home_appliances, labels = ['Electronics', 'Clothing', 'Home Appliances'])

plt.xlabel('Quaters')
plt.ylabel('Sales ($)')
plt.title('Product Category Sales Over Quaters')
plt.legend()

plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D

x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x,y = np.meshgrid(x,y)
z = np.sin(np.sqrt(x**2 + y**2))

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(x, y, z, c=z, cmap='viridis')
ax.set_xlabel('X axis')
ax.set_ylabel('Y axis')
ax.set_zlabel('Z Axis')
plt.show()
```

```python
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.tan(x)
y4 = np.exp(-x)

fig, ax = plt.subplots(2,2, figsize=(10,8))

ax[0,0].plot(x, y1, 'r')
ax[0,0].set_title('Sine wave')
ax[0,1].plot(x, y2, 'b')
ax[0,1].set_title('Cosine Wave')
ax[1,0].plot(x, y3, 'g')
ax[1,0].set_title('Tangent Wave')
ax[1,1].plot(x, y4, 'm')
ax[1,1].set_title('Exponential Decay')

plt.tight_layout()
plt.show()
```
