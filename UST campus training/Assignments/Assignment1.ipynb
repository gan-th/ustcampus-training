```python
matrix = [
    [1,2,3,4],
    [6,8,3,4],
    [7,8,9,3]
]

newmatrix = []
for i in range(4):
    l = []
    for j in range(3):
        l.append(matrix[j][i])
    newmatrix.append(l)
for i in newmatrix:
    print(i)
```

```python
import numpy as np
matrixx = np.array(matrix)
numpymatrix = matrixx.T
print(numpymatrix)
```
