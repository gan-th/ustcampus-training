```python
import pandas as pd
import random


columns = [
    "Platform Name", "Courses", "Free/Paid", "Duration",
    "Marketing - Cost per lead", "Marketing - Conversion Rate", "Marketing - Student Enrolled", "Marketing - Marketing ROI",
    "Services - Course Completion Rate", "Services - Engagement Metrics", "Services - Placement Rate", "Services - Course Variety",
    "Branding - Brand Reach", "Branding - Share of Voice", "Branding - Branded Search Trends", "Branding - Referral Rate",
    "Customer Expectation - Customer Satisfaction Score", "Customer Expectation - Customer Effort Score", "Customer Expectation - Net Promoter Score"
]


platforms = ['Platform A', 'Platform B', 'Platform C', 'Platform D', 'Platform E', 'Platform F', 'Platform G', 'Platform H', 'Platform I', 'Platform J']

courses = ['Data Science', 'Web Development', 'AI & ML', 'Cloud Computing', 'Digital Marketing', 'Cyber Security', 'Blockchain', 'UX/UI Design']


def generate_random_data():
    return {
        "Platform Name": random.choice(platforms),
        "Courses": random.choice(courses),
        "Free/Paid": random.choice(['Free', 'Paid']),
        "Duration": random.choice(['4 weeks', '6 weeks', '8 weeks', '10 weeks', '12 weeks']),
        "Marketing - Cost per lead": round(random.uniform(30, 150), 2),
        "Marketing - Conversion Rate": round(random.uniform(2, 10), 2),
        "Marketing - Student Enrolled": random.randint(50, 300),
        "Marketing - Marketing ROI": round(random.uniform(2.0, 5.0), 2),
        "Services - Course Completion Rate": random.randint(60, 95),
        "Services - Engagement Metrics": random.choice(['Low', 'Medium', 'High']),
        "Services - Placement Rate": random.randint(60, 100),
        "Services - Course Variety": random.randint(3, 8),
        "Branding - Brand Reach": random.randint(50000, 250000),
        "Branding - Share of Voice": random.randint(5, 20),
        "Branding - Branded Search Trends": random.randint(30, 75),
        "Branding - Referral Rate": random.randint(10, 25),
        "Customer Expectation - Customer Satisfaction Score": round(random.uniform(4.0, 5.0), 1),
        "Customer Expectation - Customer Effort Score": random.randint(1, 5),
        "Customer Expectation - Net Promoter Score": random.randint(40, 80)
    }


data = [generate_random_data() for _ in range(50)]
df = pd.DataFrame(data, columns=columns)

file_path = 'C:/Users/Administrator/Desktop/UST campus training/platform_data.csv'
df.to_csv(file_path, index=False)

print(f"Dataset saved to {file_path}")
```

```python
plt.figure(figsize=(15, 5))

# Customer Satisfaction Score
plt.subplot(1, 3, 1)
sns.histplot(df['Customer Expectation - Customer Satisfaction Score'], kde=True, color='skyblue')
plt.title('Customer Satisfaction Score Distribution')


# Customer Effort Score
plt.subplot(1, 3, 2)
sns.histplot(df['Customer Expectation - Customer Effort Score'], kde=True, color='lightgreen')
plt.title('Customer Effort Score Distribution')

# Net Promoter Score
plt.subplot(1, 3, 3)
sns.histplot(df['Customer Expectation - Net Promoter Score'], kde=True, color='lightcoral')
plt.title('Net Promoter Score Distribution')

plt.tight_layout()
plt.show()
```

```python
# Boxplots to identify outliers and distribution characteristics
plt.figure(figsize=(15, 5))

# Customer Satisfaction Score Boxplot
plt.subplot(1, 3, 1)
sns.boxplot(x=df['Customer Expectation - Customer Satisfaction Score'], color='skyblue')
plt.title('Customer Satisfaction Score Boxplot')

# Customer Effort Score Boxplot
plt.subplot(1, 3, 2)
sns.boxplot(x=df['Customer Expectation - Customer Effort Score'], color='lightgreen')
plt.title('Customer Effort Score Boxplot')

# Net Promoter Score Boxplot
plt.subplot(1, 3, 3)
sns.boxplot(x=df['Customer Expectation - Net Promoter Score'], color='lightcoral')
plt.title('Net Promoter Score Boxplot')

plt.tight_layout()
plt.show()
```

```python
# Scatterplots to visualize relationships between scores
plt.figure(figsize=(12, 6))

# Satisfaction vs Effort Score
plt.subplot(1, 2, 1)
sns.scatterplot(x=df['Customer Expectation - Customer Satisfaction Score'],
                y=df['Customer Expectation - Customer Effort Score'],
                color='skyblue')
plt.title('Customer Satisfaction vs Effort Score')
plt.xlabel('Satisfaction Score')
plt.ylabel('Effort Score')

# Satisfaction vs NPS
plt.subplot(1, 2, 2)
sns.scatterplot(x=df['Customer Expectation - Customer Satisfaction Score'],
                y=df['Customer Expectation - Net Promoter Score'],
                color='lightcoral')
plt.title('Customer Satisfaction vs NPS')
plt.xlabel('Satisfaction Score')
plt.ylabel('Net Promoter Score')

plt.tight_layout()
plt.show()
```

```python

# Additional Analysis: Check if there are any patterns with 'Free/Paid' courses
grouped_by_paid_status = df.groupby('Free/Paid').agg({
    'Customer Expectation - Customer Satisfaction Score': 'mean',
    'Customer Expectation - Customer Effort Score': 'mean',
    'Customer Expectation - Net Promoter Score': 'mean'
})

print("\nAverage Customer Expectation by Free/Paid:\n", grouped_by_paid_status)

# Correlation between customer satisfaction and NPS to see if there is a positive relationship
sns.lmplot(x='Customer Expectation - Customer Satisfaction Score',
           y='Customer Expectation - Net Promoter Score',
           data=df, aspect=1.5, height=6)
plt.title('Relationship Between Satisfaction and Net Promoter Score')
plt.xlabel('Customer Satisfaction Score')
plt.ylabel('Net Promoter Score')
plt.show()
```

```python
# Boxplot to compare Customer Satisfaction based on Free/Paid
plt.figure(figsize=(12, 6))
sns.boxplot(x='Free/Paid', y='Customer Expectation - Customer Satisfaction Score', data=df, palette="Set2")
plt.title('Customer Satisfaction Score by Free/Paid')
plt.xlabel('Free/Paid')
plt.ylabel('Customer Satisfaction Score')
plt.show()
```

```python
# Boxplot to compare Customer Effort Score based on Free/Paid
plt.figure(figsize=(12, 6))
sns.boxplot(x='Free/Paid', y='Customer Expectation - Customer Effort Score', data=df, palette="Set2")
plt.title('Customer Effort Score by Free/Paid')
plt.xlabel('Free/Paid')
plt.ylabel('Customer Effort Score')
plt.show()
```

```python

# Boxplot to compare Net Promoter Score based on Free/Paid
plt.figure(figsize=(12, 6))
sns.boxplot(x='Free/Paid', y='Customer Expectation - Net Promoter Score', data=df, palette="Set2")
plt.title('Net Promoter Score by Free/Paid')
plt.xlabel('Free/Paid')
plt.ylabel('Net Promoter Score')
plt.show()
```

```python
# Bar chart to show the average customer expectations for Free vs Paid courses
avg_expectations = df.groupby('Free/Paid').agg({
    'Customer Expectation - Customer Satisfaction Score': 'mean',
    'Customer Expectation - Customer Effort Score': 'mean',
    'Customer Expectation - Net Promoter Score': 'mean'
}).reset_index()

# Plotting the bar chart
avg_expectations.plot(x='Free/Paid', kind='bar', figsize=(12, 6), color=['skyblue', 'lightcoral','red'])
plt.title('Average Customer Expectations by Free/Paid')
plt.ylabel('Average Score')
plt.xlabel('Free/Paid')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```
