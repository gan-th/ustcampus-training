```python
val1 = 20
val2 = 60

print('Write the prime number between {0} and {1}.'.format(val1,val2))
for num in range(val1, val2+1):
    isPrime = True
    for i in range(2,num):
        if num%i==0:
            isPrime = False
    if isPrime :
        print(num)
```

```python
# To check if all the seats are in the same compartment

x = [11,11,11,12]
flag = True
val = x[0]
for i in x:
    if i!=val:
        flag = False
        break
if flag == True:
    print('Seat allocated')
else:
    print('Seat not allocated')
    
```
